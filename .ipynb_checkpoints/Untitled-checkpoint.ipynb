```python
import cv2
```

```python
img = cv2.imread('car.jpeg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #If we write imshow(img) directly, the code will demostrate the image in BGR,
#which is in blue, green and red. This fashion, we can convert BGR to RGB, which fully demonstrates the image.
```

```python
from deepface import DeepFace
```
